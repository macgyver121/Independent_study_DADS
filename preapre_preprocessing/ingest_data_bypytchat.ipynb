## เรื่องเล่าเช้านี้ 21 มีค 66 (MorningNEWsTV3 21/3/2023)

```python
import pytchat
import pytube
import time
import pandas as pd

video_id = '6iwVKK7TMEI'  # replace VIDEO_ID_HERE with the YouTube video ID

chat = pytchat.create(video_id=video_id)

dList = []
aList = []
mList = []
eList = []

while chat.is_alive() :
    for c in chat.get().sync_items():
        print(f"{c.datetime} [{c.author.name}]- {c.message} Time:{c.elapsedTime}")
        dList.append(c.datetime)
        aList.append(c.author.name)
        mList.append(c.message)
        eList.append(c.elapsedTime)

# dictionary of lists
dict = {'dateTime': dList, 'authorName': aList, 'Message': mList, 'Time': eList}

df = pd.DataFrame(dict)
df.head()


from google.colab import files
df.to_csv("reungrao1.csv",encoding="utf_8_sig", index=False)
files.download("reungrao1.csv")
```
