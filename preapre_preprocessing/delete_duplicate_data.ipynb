# Delete Duplicate Data

```python
import pandas as pd
```

```python
df = pd.read_csv('/content/reungrao5.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
newdf = df.drop_duplicates(
    subset = ['authorName', 'Message'],
    keep = 'first').reset_index(drop = True)
```

```python
newdf.info()
```

# Save file

```python
from google.colab import files
newdf.to_csv("reungrao5_dd.csv",encoding="utf_8_sig", index=False)
files.download("reungrao5_dd.csv")
```
